```python
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates
from matplotlib.colors import LinearSegmentedColormap
import seaborn as sb
```

```python
xcarbo_data = pd.read_csv('Export_ScaleXCarbo1_2024_10_7_15_40_22_measles.csv')
xcarbo_data.head()
```

```python
xcarbo_data.Time = pd.to_datetime(xcarbo_data.Time)
xcarbo_data.Time.dtype
df = xcarbo_data['Time'].dt.date
```

```python
pH1_001_PV         = xcarbo_data['pH1_001.PV']
pH1_001_PV         = pH1_001_PV.dropna()
series_pH1_001_PV  = xcarbo_data.loc[xcarbo_data['pH1_001.PV'].notna(), 'Time']
#series_pH1_001_PV
```

```python
#pH1_001_PV - Lineplot
plt.figure(figsize=(15, 6))
sb.lineplot(x=series_pH1_001_PV, y=pH1_001_PV, label=pH1_001_PV)
plt.xlabel("Date")
plt.ylabel("pH1_001_PV")
plt.title("pH1_001_PV measles - 2024-20-09 To 2024-05-10")
plt.show()
```

```python
#pH1_001_PV - Scatterplot
plt.figure(figsize=(15, 6))
color = LinearSegmentedColormap.from_list("CustomR-O-Y", ["#ffdc6b", "#ffb84d", "#ff8e3d", "#c70000"])
sb.scatterplot(x=series_pH1_001_PV, y=pH1_001_PV, hue=pH1_001_PV, palette=color, s=100)
plt.xlabel("Date")
plt.ylabel("pH1_001_PV")
plt.title("pH1_001_PV measles - 2024-20-09 To 2024-05-10")
plt.show()
```
