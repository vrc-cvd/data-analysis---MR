```python
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates
from matplotlib.colors import LinearSegmentedColormap
from matplotlib.animation import FuncAnimation
```

```python
my_data = pd.read_csv('Export_ScaleXCarbo1_2024_10_7_15_40_22.csv')
#my_data.head()
```

```python
my_data.Time = pd.to_datetime(my_data.Time)
my_data.Time.dtype
df = my_data['Time'].dt.date  # Extraire la partie jour
```

```python
#pH1_001_PV
all_ph1 = my_data['pH1_001_PV']
all_ph1 = all_ph1.dropna()
series_ph1 = my_data.loc[my_data['pH1_001_PV'].notna(), 'Time']
```

```python
plt.figure(figsize=(20, 10))
plt.title("pH1_001_PV Data 2024-09-20 to - 2024-10-05")
plt.xlabel("Date")
plt.ylabel("pH1_001.PV")
plt.plot(series_ph1, all_ph1)
```

```python
#Temperature
all_bio_temp = my_data['BioR_temp_001_PV']
all_bio_temp = all_bio_temp.dropna()
series_temp  = my_data.loc[my_data['BioR_temp_001_PV'].notna(), 'Time']
```

```python
#Temperature_point
colors = ['yellow' if v < 25 else 'orange' if v < 37 else 'red' for v in all_bio_temp]
temperature_cmap = LinearSegmentedColormap.from_list("TempMap", ["yellow", "orange", "red"])
plt.figure(figsize = (20, 10))
scatter  = plt.scatter(series_temp, all_bio_temp, c=all_bio_temp, cmap=temperature_cmap, s=100)
colorbar = plt.colorbar(scatter)
colorbar.set_label("Temperature (°C)")
# Add colorbar
plt.title("Temperature")
plt.xlabel("Date")
plt.ylabel("BioR temperature")
plt.show()
```

```python
#Temperature_line
plt.figure(figsize = (20, 10))
plt.plot(series_temp, all_bio_temp)
#colorbar = plt.colorbar(scatter)
#colorbar.set_label("Temperature (°C)")
# Add colorbar
plt.title("Temperature")
plt.xlabel("Date")
plt.ylabel("BioR temperature")
plt.show()
```
